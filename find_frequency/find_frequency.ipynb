```python
from collections import Counter
import re

# Danh sách từ vô nghĩa (stop words)
stop_words = {
    'và', 'là', 'của', 'trong', 'một', 'các', 'những', 'đã',
    'với', 'cho', 'rằng', 'này', 'được', 'thì', 'khi', 'đó',
    'nên', 'tôi', 'bạn',  'để', 'từ', 'vì', 'ở',
    'như', 'vẫn', 'trên', 'do', 'cả', 'không', 'chỉ', 'mà', 'còn'
}

# Đọc và tách từ
words = re.findall(r'\b\w+\b', text.lower())

# Lọc từ vô nghĩa và lưu lại các từ đã bị loại
filtered_words = []
removed_words = []

for word in words:
    if word in stop_words:
        removed_words.append(word)
    else:
        filtered_words.append(word)

# Đếm tần suất từ
word_freq = Counter(filtered_words)

# In kết quả
print(f"Tổng số từ : {len(filtered_words)}\n")
print("Tần suất xuất hiện của các từ:")
for word, freq in word_freq.items():
    print(f"'{word}': {freq}")

# Hiển thị từ đã bị xóa (stop words xuất hiện trong văn bản)
if removed_words:
    removed_freq = Counter(removed_words)
    print("\nCác từ đã bị xóa (stop words xuất hiện trong văn bản):")
    for word, freq in removed_freq.items():
        print(f"'{word}': {freq}")
else:
    print("\nKhông có từ nào bị xóa.")
```

```python
print(f"\nTổng số từ trong file: {len(words)}")
print(f"Tổng số từ xuất hiện: {len(filtered_words)}\n")
print(f"Số từ khác nhau sau khi lọc: {len(word_freq)}")
```
